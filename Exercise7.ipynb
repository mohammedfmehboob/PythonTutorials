```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
the_Data = pd.read_csv("datasets/007ExerciseFile.csv")
```

```python
display(the_Data)
```

Making an informative histogram of the mutation rates. 

```python

sns.displot(the_Data, x="Mutation_rate", hue="Strain", kind="hist", 
            alpha=0.3, binwidth = 0.2,  
            palette = ["green", "red", "blue", "orange"])
```

I also created a histogram with a KDE plot to make the distribution of the data more visually appealing.

```python
sns.displot(the_Data, x="Mutation_rate", hue="Strain", kind="hist", 
            alpha=0.3, binwidth = 0.2, kde = True, 
            palette = ["green", "red", "blue", "orange"])
```

This is the KDE plot for the data.

```python
sns.displot(the_Data, x="Mutation_rate", hue="Strain", kind="kde", fill=True, alpha=0.3, legend = True, 
            palette = ["green", "red", "blue", "orange"])
```

This is the violin plot.

```python
sns.violinplot(data=the_Data, x="Strain",
               y="Mutation_rate", inner="sticks",
               palette = ["green", "red", "blue", "orange"])
```

This is the boxplot.

```python
sns.boxplot(data=the_Data, x="Strain", y="Mutation_rate",
            palette = ["green", "red", "blue", "orange"])
```

This is the boxplot with an overlaid strip chart.

```python
sns.boxplot(data=the_Data, x="Strain", y="Mutation_rate",
            palette = ["green", "red", "blue", "orange"])
sns.stripplot(data=the_Data, x="Strain", y="Mutation_rate", 
              alpha = 0.4,  
              palette = ["green", "red", "blue", "orange"])
```

The plot that I will show to Anthony Fauci is the box plot with an overlaid strip chart. This is the best one in my opinion because it easy to understand which of the strains mutates at the highest rate (Alpha) and it also shows each of the individual data points in those strains.

```python
box = sns.boxplot(data=the_Data, x="Strain", y="Mutation_rate",
            palette = ["blue", "orange", "red", "green"],
            order = ["st_gamma" , "st_delta", "st_beta",
                     "st_alpha"]
           )
for i,each_box in enumerate(box.artists):
    each_box.set_edgecolor('white')
strip_plot= sns.stripplot(data=the_Data, x="Strain", 
              y="Mutation_rate", 
              alpha = 0.4,  
              palette = ["blue", "orange", "red", "green"],
              order = ["st_gamma" , "st_delta", "st_beta",
                       "st_alpha"])
box.set_ylabel("Mutation Rate")
strip_plot.set_ylabel("Mutation Rate")
box.set_xticklabels( ("Gamma", "Delta", "Beta", "Alpha"))
strip_plot.set_xticklabels( ("Gamma", "Delta", "Beta", 
                             "Alpha"))
plt.title("Strain and Mutation Rates")


```

From the graph above, it can be seen that the Alpha strain has the fastest mutation rate (around a median of 9). Also, the variability of the data distributions of each strain are quite different from one another. Due to this, the 4 strains should be modeled seperately.
