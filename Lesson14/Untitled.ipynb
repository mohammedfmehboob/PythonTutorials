```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
from  numpy.random import multivariate_normal
mu1 = 0;
mu2 = 0;
var1 = 1;
var2 = 1;
cov = 0.9;
cov_m = [[var1, cov],
         [cov, var2]]
data = multivariate_normal([mu1, mu2], cov_m, size=1000)
plt.scatter(data[:,0], data[:,1])
```

## Description:
The code above is creating correlated data sets and plotting them in a scatterplot. The first line imports the function that will be used to create the correlated datasets. The second line creates the variable that stores the mean of the first dataset. The third line does the same as the second line, but for the second dataset. The fifth line and sixth line stores the variances of both of the respective samples. The seventh line stores the correlation between the two datasets (the relationship between them). 

After those, the covariance matrix is created. The covariance matrix is essentially the matrix that is used to control the correlation between the two datasets. This matrix consists of four numbers, the correlation between the datasets (cov), the variance of the first dataset (var1), the variance of the second dataset (var 2), and the correlation between the datasets again. 

After the covariance matrix, the datasets are created from the aforementioned variables. A list of meas, the covariance matrix, and the size of the samples are passed as paramaters. The data, after the multivariate_normal function is used, is an array that contains 2 columns with 1000 rows for each column.Then, with the scatter function, a scatterplot is made that calls each one of those datasets.
