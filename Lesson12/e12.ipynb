```python
import numpy as np
import seaborn as sns
```

```python
data_anxiety = np.load('../datasets/012_anxiety_data.npy')
data_anxiety
```

```python
sns.color_palette('flare', as_cmap = True)
sns.histplot(data_anxiety, color = 'salmon')
```

```python
sns.kdeplot(data_anxiety, fill = True, color = 'salmon', lw = 5)
```

```python
Mean = np.mean(data_anxiety)
Median = np.median(data_anxiety)
stD = np.std(data_anxiety)

print("Mean:", Mean)
print("Median", Median)
print("Standard Deviation", stD)
```

#### The standard error of the mean is the standard deviation of the sampling distribution of means. 

```python
avg_mean = 50
avg_std = 10
n_experiments = 2000
# lets simulate our sampling distribution
simulated_exp = avg_mean + (avg_std * np.random.randn(n_experiments, 1))
print("mean:", np.mean(simulated_exp), "standard deviation", np.std(simulated_exp))
sns.displot(simulated_exp, kind = 'kde', fill = True, palette = 'magma')
```

```python
mc_means = np.zeros((n_experiments, 1))
for i in range(n_experiments):
    present_mean = avg_mean + (avg_std * np.random.randn())
    mc_means[i] = (present_mean)
```

```python
sns.displot(mc_means, kind = 'hist', palette = 'magma')
```

```python
sns.displot(mc_means, kind = 'kde', palette = 'magma')
```

```python
the_mean = np.mean(mc_means)
the_std = np.std(mc_means)
print("Mean:", the_mean)
print("Standard Deviation:", the_std)
```

## In comparison to our original mean and standard deviation, the ones from our simulation are pretty close to the data mean and standard deviation (really close to correct).

### The standard deviation is extremely close to the standard error that was found in the original data set. 

```python
bonus_simulated_exp = Mean + (stD * np.random.randn(n_experiments,1))
```

```python
sns.displot(bonus_simulated_exp, kind = 'kde', palette = 'magma', fill= 'True')
print("Mean:", np.mean(bonus_simulated_exp), "standard deviation", np.std(bonus_simulated_exp))
```
