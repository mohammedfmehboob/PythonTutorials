Finding the number of standard deviations that the first hump is in comparison to the lower values (OPTIONAL)

```python
first_row = the_data[0:21].std(1)
second_row = the_data[21:42].std(1)
```

```python
total_variance = sum((first_row * first_row) + 
                     (second_row * second_row))
```

```python
std_deviation = np.sqrt(total_variance / 40)
```

```python
std_deviation
```

```python
first_mean = the_data[0:21].mean(1)
second_mean = the_data[21:42].mean(1)
```

```python
first_total = sum(first_mean)
second_total = sum(second_mean)
```

```python
(second_total- first_total) / std_deviation
```

### Found that the hump is 18 standard deviations above the lower values.
